```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
```

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: 'cuartil-' + str(x))
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
ax.patches[2].set_facecolor('red')
```

```python
values, counter = np.unique(y, return_counts = True)
print(dict(zip(values, counter)))
```

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index)
```

```python
y = df['viento']
values, freq = np.unique(y, return_counts=True)
```

```python
fig, ax = plt.subplots()
ax.hist(y, bins = 30)
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencia')
plt.axvline(np.mean(y), c='r', linestyle='--', label='Promedio')
plt.axvline(np.mean(y) + np.std(y), c='k', linestyle='--', label='+ 1-desviacion')
plt.axvline(np.mean(y) - np.std(y), c='k', linestyle='--', label='- 1-desviacion')
plt.legend()
```
