```python
import numpy as np
from numpy.random import binomial # random numbers generator
from scipy.stats import binom # binomial function
from math import factorial
import matplotlib.pyplot as plt
```

## Secuencias de eventos tipo **Bernoulli**
![](form1.png)

```python
def my_binomial(k, n, p):
    return factorial(n) / (factorial(k) * factorial(n - k)) * pow(p, k) * pow(1 - p, n - k)
```

```python
# que salgan dos caras si lanzamos 3 veces una moneda
print(my_binomial(2, 3, 1/2))
```

```python
# Definir una ditribucion binomial para 3 intentos con 1/2 de probabilidades
dist = binom(3, 1/2)
# Identificar la probabilidad de que un evento se repita dos veces en nuestra distribucion binomial
dist.pmf(2)
```

# Función de densidad acumulada
![](form2.png)

```python
# Funcion de densidad acumulada
dist.cdf(2)
```

## Simulación de secuencias con generadores aleatorios (Escuela Frecuentista)
Tienen como propósito simular muestras de datos que resultarían de muestreos en la vida real de procesis aleatorios como lanzar una moneda o un dado

```python
p = 1/2
n = 3
binomial(n, p)
```

```python
arr = []
for _ in range(100):
    arr.append(binomial(n, p))
```

```python
def plot_hist(num_trials):
    values = [0, 1, 2, 3]
    arr = []
    for _ in range(num_trials):
        arr.append(binomial(n, p))
    sim = np.unique(arr, return_counts=True)[1] / len(arr)
    teorica = [binom(n, p).pmf(k) for k in values]
    plt.bar(values, sim, color = 'red')
    plt.bar(values, teorica, alpha = 0.5, color = 'blue')
    plt.title(f'{num_trials} experimentos')
    plt.show()

plot_hist(20)
plot_hist(200)
plot_hist(200000)
```
