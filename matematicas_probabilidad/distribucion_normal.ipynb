```python
import pandas as pd
import numpy as np
from scipy.stats import norm # normal function
import matplotlib.pyplot as plt
```

# Distribución normal teórica
![](form3.png)

```python
def gaussian(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-1/2 * pow((x - mu) / sigma, 2))
```

```python
x = np.arange(-4, 4, 0.1)
```

```python
y = gaussian(x, 0.0, 1.0)
plt.plot(x, y)
```

```python
# Desplazamos la campana
y = gaussian(x, 1.0, 1.0)
plt.plot(x, y)
```

```python
# Hacemos mas estrecha la campana
y = gaussian(x, 1.0, 0.5)
plt.plot(x, y)
```

```python
dist = norm(0, 1)
x = np.arange(-4, 4, 0.1)
y = [dist.pdf(value) for value in x]
```

```python
plt.plot(x, y)
```

```python
dist = norm(0, 1)
x = np.arange(-4, 4, 0.1)
y = [dist.cdf(value) for value in x]
plt.plot(x, y)
```

## Housefly wins gaussian distribution

```python
df = pd.read_excel('gaussian.xls')
```

```python
arr = df['Normally Distributed Housefly Wing Lengths'].values[3:]
arr
```

```python
unique_values, unique_counts = np.unique(arr, return_counts=True)
plt.bar(unique_values, unique_counts)
```

```python
mu = arr.mean()
sigma = arr.std()
dist = norm(mu, sigma)
x = np.arange(30, 60, 0.1)
y = [dist.pdf(value) for value in x]
plt.plot(x, y, color='green')
plt.bar(unique_values, unique_counts/len(arr))
```
